```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import random
import math
from matplotlib import interactive
interactive(True)
```

```python
prim = primes_method2(1000)
print(prim)
```

```python
breakDown = 3
breakDownList = prim[:breakDown]
print(breakDownList)
```

```python
def getCoordsForNum(x : int):
  coords = []
  for i in range(len(breakDownList)-1, -1, -1):
    coords.append(math.floor(x / breakDownList[i]))
    x -= coords[-1] * breakDownList[i]
  return coords[::-1]
```

```python
# 1 2 3 -> x y z
print(getCoordsForNum(4))
```

```python
numbers = list(range(1, 10000))
```

```python
primeCoords = []
for prime in prim:
  primeCoords.append(getCoordsForNum(prime))
```

```python
sequence_containing_x_vals = []
sequence_containing_y_vals = []
sequence_containing_z_vals = []

for x,y,z in primeCoords:
  sequence_containing_x_vals.append(x)
  sequence_containing_y_vals.append(z)
  sequence_containing_z_vals.append(y)
```

```python

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')


ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)
ax.plot(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)
plt.show()
```
