# **Prime 3d distribution**

```python
def primes_method2(n):
   odds = range(3, n+1, 2)
   sieve = set(sum([list(range(q*q, n+1, q+q)) for q in odds], []))
   return [2] + [p for p in odds if p not in sieve]
```

```python
prim = primes_method2(1000)
print(prim)
```

```python
breakDown = 3
breakDownList = prim[:breakDown]
print(breakDownList)
```

```python
def getCoordsForNum(x : int):
  coords = []
  for i in range(len(breakDownList)-1, -1, -1):
    coords.append(math.floor(x / breakDownList[i]))
    x -= coords[-1] * breakDownList[i]
  return coords[::-1]
```

```python
# 1 2 3 -> x y z
print(getCoordsForNum(4))
```

```python
numbers = list(range(1, 10000))
```

```python
primeCoords = []
for prime in prim:
  primeCoords.append(getCoordsForNum(prime))
```

```python
sequence_containing_x_vals = []
sequence_containing_y_vals = []
sequence_containing_z_vals = []

for x,y,z in primeCoords:
  sequence_containing_x_vals.append(x)
  sequence_containing_y_vals.append(z)
  sequence_containing_z_vals.append(y)
```

```python

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')


ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)
ax.plot(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)
plt.show()
```

# **Prime Probability**

```python
def primes(n): # simple sieve of multiples
   odds = range(3, n+1, 2)
   sieve = set(sum([list(range(q*q, n+1, q+q)) for q in odds], []))
   return [2] + [p for p in odds if p not in sieve]
```

```python
# aproximation of prime numbers count
def f(x):
  return x/np.log(x)
```

```python
# Primes
n = 1000
primesUpTo = []

for i in range(1, n+1):
  primesUpTo.append(len(primes(i)))
```

```python
x = np.linspace(1, n, n)
```

```python
fig, ax = plt.subplots()
ax.plot(x, primesUpTo)
ax.plot(x, f(x), color='red')
#ax.plot(x, x, color='green')
plt.show()
```
